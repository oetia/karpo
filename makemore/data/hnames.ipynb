```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
import numpy as np
```

## Manual

```python
def get(url: str = "https://www.animecharactersdatabase.com/sort_characters.php", 
        x: int = 0) -> requests.Response:
    querystring = {
        "x": f"{x}",
        "hair_color": "",
        "hair_length": "0",
        "sex": "2",
        "mimikko": "",
        "age": "0",
        "eye_color": "0",
        "r_eye_color": "",
        "r_hair_color": "",
        "r_hair_length": "",
        "r_sex": "",
        "r_mimikko": "",
        "r_age": "",
        "clothing": "",
        "lightdark": "",
        "otherchar": "",
        "role": "",
        "refs": "",
        "released": "",
        "display": "",
        "max": "100",
        "mode": "",
        "filter": "",
        "pid": "",
        "pids": "",
        "mt": "1",
        "s7": "",
        "order": "romjiasc",
    }

    payload = ""
    headers = {"User-Agent": "Mozilla/4.73 [en] (X11; U; Linux 2.2.15 i686)"}
    response = requests.request(
        "GET", url, data=payload, headers=headers, params=querystring
    )

    return response

get(x=0).text
```

```python
with open("names.txt", "a+") as file:
    for i in range(2, 314):
        x = i * 100
        
        response = get(x=x)
        soup = BeautifulSoup(response.text, "html.parser")
        
        divs = soup.findAll("div", attrs={"class": "tile3top"})
        page_names = list(map(lambda div: div.text, divs))
        string = "\n" + "\n".join(page_names)
        file.write(string)
        
        print(page_names[-1])
        
        time.sleep(1)
```

```python
names = np.array(open("names.txt").readlines())
names = names.reshape((names.shape[0], 1))
names.shape
```

```python
df_names = pd.DataFrame(data=names, columns=['name'])
df_names.shape
```

```python
df_names["name"] = df_names["name"].str.lower().str.strip()

df_names.drop_duplicates(inplace=True)

misc_mask = df_names["name"].str.contains("[^a-z\ \-]", regex=True)
df_names = df_names[~misc_mask]
print(df_names.shape)
```

```python
df_names.to_csv("hnames.csv", index=False, header=False)
```
